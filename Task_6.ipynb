##4.  Write a Python program to plot two or more lines with legends, different widths and colors. The code snippet gives the output shown in the following screenshot:

##6.  Write a Python programming to display a bar chart of the popularity of programming Languages.
```
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
The code snippet gives the output shown in the following screenshot:
```

image.png

````python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([0.4,1.4,2.4,3.4,4.4,5.4])
y=np.array([ 22.2, 17.6, 8.8, 8, 7.7, 6.7])
plt.bar(x,y,color='b',edgecolor='black')

plt.xlim(0,6)
plt.ylim(0,25)

x_values=[0,1,2,3,4,5]
x_labels=['Java','Python','PHP','JavaScript','C#','C++']
plt.xticks(x_values,x_labels)

plt.xlabel('Languages')
plt.ylabel('Popularity')
plt.title('Popularity of Programming Language \nWorldwide, Oct 2017 compared to a year ago')

plt.minorticks_on()
plt.grid(color='r')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')


plt.show()
````

##7.  Write a Python programming to display a horizontal bar chart of the popularity of programming Languages.
```
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
```

##Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

The Pie chart should look like this.



image.png

````python
import matplotlib.pyplot as plt
import pandas as pd

month_list=df['month_number'].tolist()

sales_data=[df['facecream'].sum(),df['facewash'].sum(),df['toothpaste'].sum(),df['bathingsoap'].sum(),df['shampoo'].sum(),df['moisturizer'].sum()]
labels=['Facecream','Facewash','Toothpaste','Bathingsoap','Shampoo','Moisturizer']

plt.pie(sales_data,labels=labels,autopct='%.1f%%')
plt.legend(loc='lower right',bbox_to_anchor=(1.60,0.05))
plt.show()
````

##Exercise Question 10: Read all product sales data and show it using the stack plot
The Stack plot should look like this.

image.png


````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

x=df['month_number']

plt.plot([],[],color='darkviolet',label='Face Cream',lw=5)
plt.plot([],[],color='cyan',label='Face wash',lw=5)
plt.plot([],[],color='red',label='Tooth paste',lw=5)
plt.plot([],[],color='black',label='Bathing soap',lw=5)
plt.plot([],[],color='green',label='Shampoo',lw=5)
plt.plot([],[],color='Olive',label='Moisturizer',lw=5)


plt.stackplot(x,df['facecream'],df['facewash'],df['toothpaste'],df['bathingsoap'],df['shampoo'],df['moisturizer'],colors=['darkviolet','cyan',
              'red','black','green','olive'])

plt.xlabel('Month Number')
plt.ylabel('Sales unit in Number')
plt.title('All product sales data using stackplot')

plt.xticks=np.arange(0,12,2)
plt.legend(loc='upper left')
plt.show()
````
